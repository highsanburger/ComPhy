```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
import sympy as sy
from scipy.linalg import eigh_tridiagonal

from sympy import var
from sympy import sympify
```

```python
    # Setting up the Grid
n = int(input("Enter n :"))
dy = 1/n
y = np.linspace(0,1,n+1)
```

```python
# # Potential function
# def ml2v(y):
#     t = y*y
#     return t
# V = ml2v(y)

# x = var('x')
# user_input = '0.01*exp((x-0.5)**2) ' #input("Enter Function - ")
# expr = sympify(user_input)
# r=[]
# for i in y:
#     res = expr.subs(x,i)
#     r.append(res)
m = float(input("Center = "))
s = float(input("Standard Dev = "))

V = -gaussian(y,m,s)+np.sin(y)
#V =1000*np.sin(20*y) * y**4 
print(V)  # -1.322...

# for i in range(len(V)):
#     V[i] = np.float64(V[i])


plt.plot(y,V)
plt.show()
```

```python
# setting up the eigenvalue matrix equation from finite difference
diag = V + 1/dy**2
off_diag = -1/(2*dy**2) * np.ones(len(diag)-1)
diag = np.array(diag,dtype=float)

ml2E,t = eigh_tridiagonal(diag,off_diag)
psi = t.T
```
